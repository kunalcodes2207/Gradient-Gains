```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'House Area': [6274, 6689, 4276, 5130, 6035, 2946, 4727, 6873, 3911, 5689, 5459, 4508, 3589, 7212, 7454, 3366],
    'Number of rooms': [8, 5, 8, 9, 6, 1, 10, 7, 4, 5, 5, 4, 8, 3, 9, 6],
    'Price': [2100, 2200, 1400, 1700, 2000, 1000, 1600, 2300, 1300, 1900, 1800, 1500, 1200, 2400, 2500, 1100]
}

df = pd.DataFrame(data)
df.head()
```

```python
X = df[['House Area', 'Number of rooms']].values
y = df['Price'].values.reshape(-1, 1)

X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X_scaled = (X - X_mean) / X_std

print("Data normalized. Mean and Standard Deviation calculated for features.")
```

```python
class MultipleLinearRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.lr = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None
        self.cost_history = []

    def compute_cost(self, X, y, y_pred):
        m = len(y)
        return (1 / (2 * m)) * np.sum(np.square(y_pred - y))

    def fit(self, X, y):
        m, n = X.shape
        self.weights = np.zeros((n, 1))
        self.bias = 0

        for i in range(self.iterations):
            y_pred = np.dot(X, self.weights) + self.bias
            
            cost = self.compute_cost(X, y, y_pred)
            self.cost_history.append(cost)

            dw = (1 / m) * np.dot(X.T, (y_pred - y))
            db = (1 / m) * np.sum(y_pred - y)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

model = MultipleLinearRegression(learning_rate=0.1, iterations=100)
model.fit(X_scaled, y)

print("Training complete.")
print(f"Final Weights: {model.weights.flatten()}")
print(f"Final Bias: {model.bias}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(model.cost_history, color='blue')
plt.title("Error Reduction During Training")
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.show()
```
